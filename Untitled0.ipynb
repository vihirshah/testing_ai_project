```python
# Path to the dataset on your Google Drive
file_path1 = '/content/gdrive/My Drive/python/climate_change_indicators.csv'

df_climate_data = pd.read_csv(file_path1)

```

```python
# Display the first few rows of the dataset to inspect it
df_climate_data.head()
```

```python
# Select only the columns related to temperature (those starting with 'F')
temperature_columns = [col for col in df_climate_data.columns if col.startswith('F')]

# Create a new DataFrame with just the 'Country' column and the temperature data
df_temperature = df_climate_data[['Country'] + temperature_columns]

# Fill missing values with 0 (or apply any other technique if needed)
df_temperature = df_temperature.fillna(0)

# Display the cleaned data
df_temperature.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact
```

```python
# Plot average temperature anomaly by year
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Average'], label='Average Temperature Anomaly')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.title('Average Global Temperature Anomaly Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Create a function to plot the monthly temperature anomaly for a selected year
def plot_monthly_data(year):
    # Filter data for the selected year
    monthly_data = df[df['Year'] == year].iloc[:, 1:-1].T  # Select month columns
    monthly_data.columns = ['Temperature Anomaly']

    # Plot the data
    plt.figure(figsize=(10, 5))
    monthly_data.plot(kind='bar', color='skyblue')
    plt.title(f'Monthly Temperature Anomaly for {year}')
    plt.xlabel('Month')
    plt.ylabel('Temperature Anomaly (C)')
    plt.xticks(rotation=45)
    plt.show()

# Create the interactive widget for selecting a year
interact(plot_monthly_data, year=(1880, 2024, 1))



```

```python


# Select only the columns for the months and year
monthly_data = df.iloc[:, 1:-1]

# Set the 'Year' column as the index
monthly_data.index = df['Year']

# Create a heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(monthly_data, cmap='coolwarm', annot=False, fmt='.2f', cbar_kws={'label': 'Temperature Anomaly (C)'})
plt.title('Monthly Global Temperature Anomalies (1880 - 2024)')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

```python
import ipywidgets as widgets
from ipywidgets import interact

```

```python
def plot_temperature_changes(country):
    # Filter the data for the selected country
    country_data = df_temperature[df_temperature['Country'] == country].iloc[0, 1:]

    # Sort the temperature data to find the top 3 largest and smallest changes
    sorted_data = country_data.sort_values()

    # Get the top 3 smallest and top 3 largest temperature changes
    top_3_smallest = sorted_data.head(3)
    top_3_largest = sorted_data.tail(3)

    # Plot the results
    plt.figure(figsize=(12, 6))

    # Plot top 3 smallest temperature changes
    plt.subplot(1, 2, 1)
    top_3_smallest.plot(kind='bar', color='red', title=f'Top 3 Smallest Temperature Changes ({country})')
    plt.xlabel('Year')
    plt.ylabel('Temperature Change (C)')

    # Plot top 3 largest temperature changes
    plt.subplot(1, 2, 2)
    top_3_largest.plot(kind='bar', color='green', title=f'Top 3 Largest Temperature Changes ({country})')
    plt.xlabel('Year')
    plt.ylabel('Temperature Change (C)')

    plt.tight_layout()
    plt.show()
```

```python
# Create a dropdown widget for selecting a country
country_widget = widgets.Dropdown(
    options=df_climate_data['Country'].unique(),  # Get unique countries from the dataset
    description='Country:',
    disabled=False
)

# Use the interact function to link the dropdown to the plotting function
interact(plot_temperature_changes, country=country_widget)
```

```python
# Melt the DataFrame to create a long format suitable for plotting
df_long = pd.melt(df_climate_data,
                  id_vars=['Country'],
                  var_name='Year',
                  value_name='Temperature Change')

# Extract the year as a number and handle invalid years gracefully
df_long['Year'] = df_long['Year'].str.extract('(\d+)').astype(float)  # Convert to float first
df_long['Temperature Change'] = pd.to_numeric(df_long['Temperature Change'], errors='coerce')

# Remove rows where 'Year' or 'Temperature Change' is NaN
df_long = df_long.dropna(subset=['Year', 'Temperature Change'])

# Convert 'Year' to integer after removing NaN values
df_long['Year'] = df_long['Year'].astype(int)

# Check the cleaned data
df_long.head()
```

```python
# Function to plot the Surface Temperature Change for a selected country
def plot_temperature_change(country):
    # Filter data for the selected country
    country_data = df_long[df_long['Country'] == country]

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(country_data['Year'], country_data['Temperature Change'], marker='o')

    # Set plot labels and title
    plt.title(f"Surface Temperature Change for {country}", fontsize=16)
    plt.xlabel("Year", fontsize=12)
    plt.ylabel("Temperature Change (C)", fontsize=12)
    plt.grid(True)
    plt.xticks(rotation=45)

    # Show the plot
    plt.show()
```

```python
# Create a dropdown widget for country selection
country_dropdown = widgets.Dropdown(
    options=df_climate_data['Country'].unique(),
    description='Country:',
    disabled=False
)

# Use the `interact` function to make the plot interactive
interact(plot_temperature_change, country=country_dropdown)
```
